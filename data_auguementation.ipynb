```python
data1=data1/255.0
data1.max()
```

```python
datagen = ImageDataGenerator(
                            rotation_range=40,
                            width_shift_range=0.2,
                            height_shift_range=0.2,
                            shear_range=0.2,
                            zoom_range=0.2,
                            horizontal_flip=True,
                            fill_mode='nearest')
```

```python
data1 = data1.reshape((1,) + data1.shape)
```

 when ever the datagen.flow(data1) then only one image is formed with some changes in it and to get a new type of image 
then you have to run the datagen.flow(data1) again.
That's why the output of datagen.flow(data1) is a NumpyArrayIterator which means whenever we perform the function a new 
image is generated

```python
test=datagen.flow(data1)
```

```python
i = 0

for batch in datagen.flow(data1, save_prefix='test', save_format='jpeg'):  # this loops runs forever until we break, saving images to current directory with specified prefix
    plt.figure(i)
    plot = plt.imshow(image.img_to_array(batch[0]))
    i += 1
    if i > 4:  # show 4 images
        break

plt.show()
```
